```python
# Load library
import pandas as pd
import country_converter as coco
```

```python
# Load country converter
converter = coco.CountryConverter()

# Create URL to JSON file
url = 'https://gist.githubusercontent.com/sindresorhus/1341699/raw/84704529d9ee4965df2cddc55e5f2bc3dc686950/countrycode-latlong-array.json'

# Load the JSON file into a data frame
countries_df = pd.read_json(url).transpose()
countries_df.drop(['an','ap','eu'], inplace=True)
countries_df = countries_df.reset_index()

# View the first ten rows
countries_df.head(10)

```

```python
codes = countries_df['index'].tolist()

print(codes)
```

```python
names = converter.convert(names = codes, src='ISO2', to='name_short')
print(names)
```

```python
names_df = pd.DataFrame(names) 
names_df.columns = ['Countries']
names_df
```

```python
newnames_df = pd.concat([countries_df, names_df], axis=1)
newnames_df
```

```python
newnames_df = newnames_df[['Countries', 0, 1]]
newnames_df
```
